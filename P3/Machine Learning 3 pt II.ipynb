<h4><> Proses Training Model Machine Learning Secara Umum</h4>
<p>
    Dalam <em>machine learning</em>, proses <em>training</em> adalah salah satu langkah terpenting untuk membangun model yang dapat membuat prediksi yang akurat. <em>Training</em> adalah tahap di mana model belajar dari data untuk mengenali pola-pola yang ada sehingga dapat digunakan untuk melakukan klasifikasi, prediksi, atau estimasi pada data yang belum pernah dilihat sebelumnya. Algoritma yang dilatih pada data ini akan menyesuaikan parameter-parameternya secara bertahap hingga mencapai hasil yang optimal.
</p>
<p>
    Setiap model <em>machine learning</em> memerlukan dua hal utama dalam proses <em>training</em>: data yang digunakan untuk melatih model, dan data yang menjadi target atau label dari proses tersebut. Setelah model dilatih, model tersebut bisa digunakan untuk memprediksi hasil dari data baru yang tidak diketahui labelnya.
</p>
<p>
    Selanjutnya, kita akan melihat tahapan proses <em>training</em> model secara umum yang dijelaskan dalam gambar.

<h5>Langkah-langkah Proses Training Model</h5>

<p><strong>1. Training Set</strong></p>
<p>
    Proses dimulai dengan <em>training set</em>, yang terdiri dari dua bagian:
    <ul>
        <li><strong>X_train</strong>: Fitur-fitur input dari data. Fitur-fitur ini bisa berupa berbagai jenis data seperti angka, kategori, atau teks yang merepresentasikan karakteristik objek yang ingin kita pelajari.</li>
        <li><strong>y_train</strong>: Label atau target yang ingin diprediksi oleh model. Misalnya, dalam klasifikasi buah, <em>y_train</em> mungkin berupa jenis buah seperti <em>apple</em>, <em>grape</em>, atau <em>lemon</em>.</li>
    </ul>
    Data ini merupakan dasar yang digunakan oleh algoritma <em>machine learning</em> untuk mengenali pola dan hubungan antara fitur dan label.
</p>

<p><strong>2. Model</strong></p>
<p>
    Setelah dataset disiapkan, <em>X_train</em> dan <em>y_train</em> dimasukkan ke dalam algoritma <em>machine learning</em> untuk membangun model. Model ini mencoba belajar dari pola dalam data pelatihan untuk menemukan cara terbaik dalam mengklasifikasikan atau memprediksi hasil yang benar berdasarkan masukan fitur.
</p>
<p>
    Algoritma ini melalui iterasi proses pembelajaran, di mana ia memperbaiki asumsi dan hubungannya berdasarkan kesalahan yang ditemukan selama latihan.
</p>

<p><strong>3. Trained Model</strong></p>
<p>
    Setelah proses pelatihan selesai, hasilnya adalah model yang telah dilatih, disebut <strong>trained model</strong>. Model ini berisi semua informasi yang telah dipelajari dari dataset pelatihan dan siap untuk digunakan.
</p>
<p>
    Model ini sekarang mampu memprediksi atau mengklasifikasikan data baru berdasarkan pola-pola yang telah dipelajari.
</p>

<p><strong>4. X_new</strong></p>
<p>
    Pada tahap selanjutnya, model yang sudah dilatih digunakan untuk memprediksi data baru yang belum pernah dilihat sebelumnya. Data baru ini disebut <strong>X_new</strong>.
</p>
<p>
    Misalnya, jika model tersebut digunakan untuk memprediksi jenis buah berdasarkan ukuran dan warna, maka <em>X_new</em> bisa berisi ukuran dan warna buah yang tidak diketahui jenisnya.
</p>

<p><strong>5. y_pred</strong></p>
<p>
    Hasil dari prediksi model pada data baru ini disebut <strong>y_pred</strong>. Ini adalah hasil prediksi dari model yang sudah dilatih sebelumnya. Jika modelnya akurat, maka nilai <em>y_pred</em> akan mendekati atau sama dengan label yang benar untuk data baru tersebut.
</p>
<p>
    Sebagai contoh, jika model dilatih untuk mengklasifikasikan jenis buah, maka <em>y_pred</em> bisa berupa prediksi bahwa buah baru adalah <em>apple</em>, <em>grape</em>, atau <em>lemon</em>.
</p>

<p>
    Dengan memahami langkah-langkah ini, kita bisa melihat bagaimana model <em>machine learning</em> dilatih dari data dan bagaimana ia digunakan untuk membuat prediksi pada data baru. Proses ini merupakan dasar dari berbagai jenis algoritma, termasuk <em>Random Forest</em>, yang menggunakan banyak pohon keputusan untuk memperkuat akurasi prediksi.
</p>


<h4><> Pengenalan Ensemble Learning</h4>
<strong> Ensemble Learning: heterogeneous & homogeneous</strong>
 <p>Referensi: <a href="https://en.wikipedia.org/wiki/Ensemble_learning">https://en.wikipedia.org/wiki/Ensemble_learning</a></p>

<p><strong>Ensemble Learning</strong> adalah teknik dalam machine learning di mana beberapa model digabungkan untuk melakukan prediksi secara lebih akurat daripada model individual. Tujuannya adalah untuk mengurangi kesalahan prediksi dan meningkatkan performa model secara keseluruhan. Dengan menggabungkan kekuatan dari beberapa model, ensemble dapat mengurangi variabilitas (variance), bias, atau ketidakseimbangan yang mungkin ada pada model tunggal.</p>


<img src="3b2.png" style="max-width: 600px;" alt="">
<h5>Langkah-langkah Ensemble Learning:</h5>

<ol>
  <li><strong>Training Set</strong><br>
      Proses dimulai dengan sebuah training set yang berisi fitur input (X_train) dan label target (y_train). Sama seperti pada model machine learning standar, dataset ini digunakan untuk melatih model agar dapat memprediksi output yang diinginkan.</li>
  
  <li><strong>Training Multiple Models</strong><br>
      Dalam ensemble learning, training set yang sama digunakan untuk melatih beberapa model. Pada gambar di atas, tiga jenis model digunakan:
      <ul>
        <li><strong>KNN (K-Nearest Neighbors)</strong></li>
        <li><strong>SVM (Support Vector Machine)</strong></li>
        <li><strong>Decision Tree</strong></li>
      </ul>
      Setiap model ini akan belajar dari data yang sama tetapi dengan algoritma yang berbeda. Karena algoritma yang berbeda memiliki cara yang berbeda dalam menemukan pola dari data, mereka menghasilkan prediksi yang berbeda.</li>

  <li><strong>Trained Models</strong><br>
      Setelah proses training, setiap model menghasilkan trained model masing-masing, yang sudah "belajar" dari data. Model yang terlatih ini siap untuk memprediksi hasil berdasarkan data baru (X_new).</li>

  <li><strong>Prediksi Model</strong><br>
      Setiap trained model akan membuat prediksi pada dataset baru (X_new). Pada gambar, kita melihat tiga prediksi yang dihasilkan:
      <ul>
        <li><strong>y_pred 1</strong> dari model KNN</li>
        <li><strong>y_pred 2</strong> dari model SVM</li>
        <li><strong>y_pred 3</strong> dari model Decision Tree</li>
      </ul>
  </li>

  <li><strong>Mean or Mode (Penggabungan Prediksi)</strong><br>
      Karena ketiga model memberikan hasil prediksi yang berbeda-beda, kita perlu menggabungkannya untuk mendapatkan prediksi final. Teknik penggabungan yang paling umum digunakan dalam ensemble learning adalah dengan menghitung rata-rata (mean) atau modus (mode) dari hasil prediksi. Dengan cara ini, kita bisa mendapatkan hasil prediksi yang lebih stabil dan akurat.</li>

  <li><strong>y_pred final</strong><br>
      Hasil akhir dari prediksi gabungan disebut sebagai <strong>y_pred final</strong>. Ini adalah prediksi yang dihasilkan dari penggabungan output tiga model yang berbeda, dan dianggap lebih akurat karena menggabungkan informasi dari berbagai perspektif algoritma.</li>
</ol>

<h5>Manfaat Ensemble Learning</h5>

<ul>
  <li><strong>Mengurangi Varians dan Bias:</strong> Dengan menggabungkan beberapa model, ensemble dapat mengurangi kesalahan yang mungkin disebabkan oleh variabilitas data atau bias pada satu model.</li>
  <li><strong>Meningkatkan Akurasi:</strong> Hasil prediksi dari beberapa model sering kali lebih baik daripada menggunakan model tunggal.</li>
  <li><strong>Stabilitas Model:</strong> Ensemble learning membuat model lebih tahan terhadap perubahan kecil pada dataset, yang dikenal sebagai robustness.</li>
</ul>

<p>Metode ensemble learning ini sering digunakan dalam berbagai aplikasi, termasuk classification dan regression, untuk mendapatkan hasil yang lebih baik dan lebih andal.</p>


<h4><> Pengenalan Bootstrap Aggregating | Bagging</h4>
<strong> Bagging: Boostrap Aggregating</strong>
 <p>Referensi: <a href="https://en.wikipedia.org/wiki/Bootstrap_aggregating">https://en.wikipedia.org/wiki/Bootstrap_aggregating</a></p>
<img src="3b3.png" style="max-width: 600px;" alt="">

<p>
    Bagging, singkatan dari Bootstrap Aggregating, adalah metode ensemble learning yang menggabungkan beberapa model 
    prediksi untuk meningkatkan akurasi prediksi dan mengurangi varians. Prinsipnya adalah dengan membangun banyak 
    model dengan data bootstrap dari dataset asli, kemudian menggabungkan prediksi mereka untuk mendapatkan prediksi akhir.
</p>

<h5>Bagaimana cara kerja Bagging?</h5>

<ol>
    <li>
        <strong>Bootstrap Sampling:</strong> Dataset asli diambil secara acak dengan pengembalian, menciptakan dataset 
        baru yang disebut bootstrap sample. Setiap bootstrap sample memiliki ukuran yang sama dengan dataset asli, 
        tetapi beberapa data mungkin muncul lebih dari sekali, sementara yang lain mungkin tidak muncul sama sekali.
    </li>
    <li>
        <strong>Membangun Model:</strong> Model prediksi (misalnya, pohon keputusan, regresi linier) dilatih pada setiap 
        bootstrap sample yang berbeda. Ini menghasilkan sejumlah model yang independen satu sama lain.
    </li>
    <li>
        <strong>Menggabungkan Prediksi:</strong> Untuk mendapatkan prediksi akhir, prediksi dari semua model yang 
        dilatih dikombinasikan. Cara penggabungannya tergantung pada jenis masalah:
        <ul>
            <li>Untuk klasifikasi, prediksi dari semua model dikumpulkan dan prediksi kelas dengan jumlah suara terbanyak 
            dipilih.</li>
            <li>Untuk regresi, prediksi dari semua model dirata-ratakan.</li>
        </ul>
    </li>
</ol>

<h3>Proses Bagging</h3>
<p>
    Gambar di bawah ini menggambarkan proses Bagging (Bootstrap Aggregating) untuk membangun model ensemble pada 
    Random Forest.
</p>

<img src="3b3.png" style="max-width: 600px;" alt="">

<div>
    <strong>1. Training Set:</strong> Dimulai dengan dataset pelatihan asli (X_train, y_train).
</div>
<div>
    <strong>2. Random Sampling with Replacement:</strong> Dataset asli diambil sampelnya secara acak dengan 
    penggantian (bootstrap) untuk menciptakan beberapa kumpulan dataset baru yang disebut "Bag". Setiap Bag berisi 
    data yang dipilih secara acak dengan kemungkinan data yang sama untuk dipilih berulang kali.
</div>
<div>
    <strong>3. Model Training:</strong> Model yang sama (dalam kasus ini, "Model") dilatih pada setiap Bag yang berbeda.
</div>
<div>
    <strong>4. Prediction:</strong> Setiap model dilatih pada Bag yang berbeda kemudian menghasilkan prediksi (X_new) 
    untuk data baru.
</div>
<div>
    <strong>5. Mean or Mode:</strong> Prediksi dari semua model digabungkan menggunakan mean atau mode (tergantung 
    pada jenis tugas) untuk mendapatkan prediksi akhir (y_pred).
</div>


<h4><> Pengenalan Random Forest | Hutan Acak</h4>
<Strong>Random Forest</Strong>
 <p>Referensi: <a href="https://en.wikipedia.org/wiki/Random_forest">https://en.wikipedia.org/wiki/Random_forest</a></p>

<h5>Mengenal Random Forest: Hutan Acak untuk Prediksi yang Lebih Akurat</h5>

<p>
    Random Forest adalah teknik pembelajaran mesin yang memanfaatkan kekuatan banyak decision tree. Bayangkan 
    memiliki sekelompok pohon yang masing-masing membuat prediksi berdasarkan data yang berbeda, dan kemudian 
    "menghitung suara" untuk menentukan jawaban akhir. Inilah konsep dasar dari Random Forest.
</p>

<img src="3b4.png" style="max-width: 600px;" alt="">
<p><strong>Bagaimana Random Forest Bekerja?</strong></p>

<ol>
    <li>
        <p><strong>Pembentukan Bagging (Bootstrap Aggregating):</strong></p>
        <ul>
            <li>Data pelatihan dibagi menjadi beberapa subset (bag) secara acak. Setiap bag mungkin berisi beberapa 
            data yang sama, seperti mengambil sampel dengan penggantian.</li>
            <li>Misalnya, kita punya 100 data, lalu dibagi menjadi 5 bag, setiap bag berisi 20 data.</li>
        </ul>
    </li>
    <li>
        <p><strong>Features Randomness (Acak Fitur):</strong></p>
        <ul>
            <li>Selain pembagian data, setiap decision tree dalam Random Forest juga menggunakan subset fitur secara 
            acak.</li>
            <li>Ini membantu mencegah model menjadi terlalu sensitif terhadap fitur tertentu dan meningkatkan 
            generalisasi model.</li>
            <li>Misalnya, jika ada 10 fitur dalam data, setiap decision tree hanya akan menggunakan 5 fitur secara 
            acak.</li>
        </ul>
    </li>
    <li>
        <p><strong>Pembuatan Pohon Keputusan (Decision Tree):</strong></p>
        <ul>
            <li>Setiap bag dibentuk, kemudian decision tree dibentuk untuk setiap bag.</li>
            <li>Setiap decision tree dilatih pada data dalam bag-nya dan subset fitur yang acak.</li>
            <li>Karena setiap pohon dilatih dengan data dan fitur yang berbeda, hasilnya pun akan berbeda.</li>
        </ul>
    </li>
    <li>
        <p><strong>Penggabungan Prediksi (Aggregation):</strong></p>
        <ul>
            <li>Setelah semua decision tree terlatih, model Random Forest akan menggunakan mean (rata-rata) atau 
            mode (nilai yang paling sering) dari prediksi setiap pohon untuk membuat prediksi akhir.</li>
            <li>Jika kita punya 5 decision tree, maka prediksi final adalah rata-rata dari 5 prediksi pohon.</li>
        </ul>
    </li>
</ol>

<p><strong>Keunggulan Random Forest:</strong></p>
<ul>
    <li><strong>Generalisasi yang Baik:</strong> Random Forest cenderung mengatasi masalah overfitting, yang 
    terjadi ketika model terlalu "terlatih" dan tidak dapat bekerja dengan baik pada data baru.</li>
    <li><strong>Stabil:</strong> Random Forest kurang rentan terhadap noise dan outliers dalam data.</li>
    <li><strong>Dapat Mengatasi Masalah Kompleks:</strong> Random Forest dapat menangani data dengan banyak fitur 
    dan hubungan yang kompleks di antara fitur tersebut.</li>
    <li><strong>Dapat Digunakan untuk Regresi dan Klasifikasi:</strong> Random Forest dapat digunakan untuk 
    memprediksi nilai numerik (regresi) atau kategori (klasifikasi).</li>
</ul>

<p>
    Dengan demikian, Random Forest adalah teknik yang sangat efektif untuk meningkatkan akurasi prediksi dan 
    mengurangi varians.
</p>


<h4><> Persiapan dataset: Iris Dataset</h4>
<strong>Referensi:</strong> <a href="https://en.wikipedia.org/wiki/Iris_flower_data_set" style="cursor: pointer; color: rgb(50, 134, 255); text-decoration: none;">https://en.wikipedia.org/wiki/Iris_flower_data_set</a>

```python
print ('Nama: Aditya Rimandi Putra')
print ('NPM : 41155050210030\n')

# Mengimpor fungsi load_iris dari modul sklearn.datasets untuk memuat dataset Iris
from sklearn.datasets import load_iris

# Memuat dataset Iris dan memisahkan fitur (X) dan label target (y)
# return_X_y=True mengembalikan fitur dan label sebagai dua variabel terpisah
X, y = load_iris(return_X_y=True)

# Mencetak dimensi dari fitur yang dimuat
# X.shape memberikan ukuran dari array fitur
print(f'Dimensi Feature: {X.shape}')

# Mencetak kelas yang ada dalam label target
# set(y) digunakan untuk mendapatkan himpunan unik dari kelas
print(f'Class: {set(y)}')
```

```python
print ('Nama: Aditya Rimandi Putra')
print ('NPM : 41155050210030\n')

# Mengimpor fungsi train_test_split dari modul sklearn.model_selection
from sklearn.model_selection import train_test_split

# Membagi dataset menjadi data pelatihan (X_train, y_train) dan data pengujian (X_test, y_test)
# test_size=0.3 menunjukkan bahwa 30% dari data akan digunakan sebagai data pengujian
# random_state=0 memastikan bahwa pembagian data bersifat reproduktif (hasil yang sama setiap kali dijalankan)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3, 
                                                    random_state=0)
```

<h4><> Implementasi Random Forest Classifier dengan Scikit Learn</h4>
Classification dengan Random Forest Classifier

```python
print ('Nama: Aditya Rimandi Putra')
print ('NPM : 41155050210030\n')

# Mengimpor kelas RandomForestClassifier dari modul sklearn.ensemble
from sklearn.ensemble import RandomForestClassifier

# Membuat model Random Forest dengan 100 estimator (pohon keputusan)
# n_estimators=100 menentukan jumlah pohon dalam hutan acak
# random_state=0 digunakan untuk memastikan hasil yang konsisten pada setiap run
model = RandomForestClassifier(n_estimators=100, random_state=0)

# Melatih model menggunakan data pelatihan (X_train) dan label target (y_train)
# Proses ini akan menyesuaikan model dengan pola yang ada dalam data pelatihan
model.fit(X_train, y_train)
```

<h4><> Evaluasi model dengan Classification Report</h4>

```python
print ('Nama: Aditya Rimandi Putra')
print ('NPM : 41155050210030\n')

# Mengimpor fungsi classification_report dari modul sklearn.metrics
from sklearn.metrics import classification_report

# Menggunakan model yang telah dilatih untuk memprediksi label target pada data pengujian (X_test)
y_pred = model.predict(X_test)

# Mencetak laporan klasifikasi yang menunjukkan metrik evaluasi model
# Laporan ini mencakup precision, recall, dan f1-score untuk setiap kelas
# y_test adalah label target sebenarnya, sedangkan y_pred adalah label target yang diprediksi oleh model
print(classification_report(y_test, y_pred))
```
